```python
import os
import requests
import numpy as np
```

```python
url_file = 'C:/Users/pilot/Starburst/sdss_veri_cekme/0.txt' # veri listesini ad ve bulunduu konum 
save_directory = 'C:/Users/pilot/Starburst/sdss_veri_cekme/data_file' # indirilecek klasrn ad

os.makedirs(save_directory, exist_ok=True)

urls = np.loadtxt(url_file, dtype=str)

for url in urls:
    try:
        response = requests.get(url)
        response.raise_for_status()  
        
        file_name = os.path.join(save_directory, f'{url[-25:]}')
        
        with open(file_name, 'wb') as file:
            file.write(response.content)
        
        print(f'{url} -> {file_name} olarak indirildi.')
    
    except requests.exceptions.RequestException as e:
        print(f'{url} indirilemedi: {e}')
```
